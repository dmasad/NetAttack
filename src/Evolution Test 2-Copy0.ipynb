```python
def global_reaching_centrality(graph):
    '''
    Compute Global Reach Centrality measure of heirarchy, as defined in:
        http://www.plosone.org/article/info%3Adoi%2F10.1371%2Fjournal.pone.0033799

    This implementation is for unweighted, undirected graphs, but the metric
    can be generalized to weighted and directed graphs as well.
    '''
    N = len(graph)

    # Compute all the C''_R(i) values
    closeness_scores = []
    for node in graph.nodes():
        c = 0
        for alter in graph.nodes():
            if node != alter:
                try:
                    d = nx.shortest_path_length(graph, node, alter)
                    c += (1.0)/d
                except:
                    pass
        c *= (1.0/(N-1))
        closeness_scores.append(c)
    
    c_max = max(closeness_scores)

    # Compute GRC:
    grc = 0
    for c in closeness_scores:
        grc += abs(c_max - c)
    grc *= (1.0/(N-1))
    return (1 - grc)
```
