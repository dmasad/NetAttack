```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

## Load the data

```python
PATH = "/Users/dmasad/Dropbox/csss2013CoEv/Unopposed Runs/"
```

```python
defender_cols = ["Generation", "ID", "Fitness"]
defender_cols += ["Gene"+str(i) for i in range(1,11)]
```

```python
run1 = pandas.read_csv(PATH + "defenders_unopposed1.csv", skiprows=1, names=defender_cols)
run2 = pandas.read_csv(PATH + "defenders_unopposed2.csv", skiprows=1, names=defender_cols)
run3 = pandas.read_csv(PATH + "defenders_unopposed3.csv", skiprows=1, names=defender_cols)
```

## Mean fitness

```python
agg1 = run1.groupby(by=["Generation"]).aggregate([np.mean, np.max])["Fitness"]
agg1.plot()
```

```python
agg2 = run2.groupby(by=["Generation"]).aggregate([np.mean, np.max])["Fitness"]
agg2.plot()
```

```python
agg3 = run3.groupby(by=["Generation"]).aggregate([np.mean, np.max])["Fitness"]
agg3.plot()
```

# Gene means

```python
genes1 = run1.groupby(by=["Generation"]).aggregate(np.mean)
genes1[genes1.columns[2:]].plot(legend=True)
```

```python
genes2 = run2.groupby(by=["Generation"]).aggregate(np.mean)
genes2[genes2.columns[2:]].plot(legend=True)
```

```python
genes3 = run3.groupby(by=["Generation"]).aggregate(np.mean)
genes3[genes3.columns[2:]].plot(legend=True)
```
