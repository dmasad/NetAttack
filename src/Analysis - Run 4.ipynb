```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
```

## Load the data

```python
PATH = "/Users/dmasad/Dropbox/csss2013CoEv/Run 4 Data/"
```

```python
attacker_cols = ["Generation", "ID", "Fitness"]
attacker_cols += ["Gene"+str(i) for i in range(1,8)]
```

```python
defender_cols = ["Generation", "ID", "Fitness"]
defender_cols += ["Gene"+str(i) for i in range(1,11)]
```

```python
attacker_data = pandas.read_csv(PATH + "attackers.csv", skiprows=1, names=attacker_cols)
```

```python
defender_data = pandas.read_csv(PATH + "defenders.csv", skiprows=1, names=defender_cols)
```

## Mean fitness

```python
defender_mean = defender_data.groupby(by=["Generation"]).aggregate(np.mean)["Fitness"]
attacker_mean = attacker_data.groupby(by=["Generation"]).aggregate(np.mean)["Fitness"]
```

```python
fitness_df = pandas.DataFrame({"Attacker": attacker_mean, "Defender": defender_mean})
```

```python
fitness_df.plot()
```

# Gene means

```python
attacker_genes = attacker_data.groupby(by=["Generation"]).aggregate(np.mean)
```
